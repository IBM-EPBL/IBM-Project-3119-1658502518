## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
import datetime
sta=datetime.date(2023,1,1)
end=datetime.date(2023,2,11)
dell=datetime.timedelta(days=1)


for i in range((end-sta).days):
  
  print(sta+i*dell)
```
